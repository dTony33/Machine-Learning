```python
from numpy import sqrt
import math 
def distance(feature1,feature2):
  d = 0.0
  for i in range(len(feature1)-1):
    d += (feature1[i] - feature2[i])**2 
  return sqrt(d)  
```

```python
x_numpy = np.array(x_test)
x_numpy
for i in x_numpy:
  print(i)
```

```python
from scipy.stats import mode
def kmeans(x_train, y_train , x_test, k):
    labels = []
    for test_feature in x_test:
      distances = []
      for feature in range(len(x_train)):
        distances.append(np.array(distance(x_train.iloc[feature,:].values , test_feature)))   
      distances = np.argsort(distances)[:k] 
      label = mode(y_train[distances])
      labels.append(label.mode[0])

    return labels
```

```python
y_pred = kmeans(x_train,y_train,x_numpy,3)
```

```python
accuray = []
k_values = []
for i in range(1,20):
  y = kmeans(x_train,y_train,x_numpy,i)
  accuray.append(accuracy_score(y_test,y))
  k_values.append(i)
```

```python
import matplotlib.pyplot as plt
plt.bar(k_values, accuray)
```

When K values is below 10, the accuray of the model is 0.6 and if it is above 10 the accuaray is fluctuating between 0.4 and 0.6.
